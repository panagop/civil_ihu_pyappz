```python
import pandas as pd
```

## Δεδομένα

### Διάβασμα του αρχείου excel με τα δεδομένα

```python
excel_filename = r'D:\ihu\Χαρτούρα\Γνωστικά αντικείμενα\2023 Απρίλιος - Τελικά\Μητρώο Εξωτερικό 2023-04-25_copy2.xlsx'
excel_file = pd.ExcelFile(excel_filename)
```

```python
worksheet_names = excel_file.sheet_names
print(worksheet_names)
```

### Συνάρτηση για το διάβασμα των δεδομένων από κάθε worksheet

```python
def read_external_table(excel_filename: str, sheet_name: str) -> pd.DataFrame:
    df = pd.read_excel(excel_filename, sheet_name=sheet_name, usecols="B:L", skiprows=8, engine='openpyxl')
    return df
```

### Δημιουργία μιας λίστας με τα dataframes από κάθε worksheet

```python
dfs = [read_external_table(excel_filename, wn) for wn in worksheet_names]
```

## Πίνακας εκλεκτόρων

### Ενιαίο dataframe με όλα τα δεδομένα των εκλεκτόρων

```python
df = pd.concat(dfs, ignore_index=True, verify_integrity=True)
# df = pd.concat(dfs, ignore_index=True)
df.info()
```

```python
df = df.drop_duplicates(subset='Κωδικός Χρήστη', keep="first")
df = df.drop(['Χαρακτη-ρισμός'], axis=1)
df.info()
```

### Αποθήκευση του φύλλου με τους εκλέκτορες σε αρχείο excel

```python
df.to_excel('eklektores.xlsx')
```

## Πίνακας γνωστικών αντικειμένων

```python
def read_external_antikeimena(excel_filename: str, sheet_name: str) -> pd.DataFrame:
    df = pd.read_excel(excel_filename, sheet_name=sheet_name, usecols="E:F, K", skiprows=4, engine='openpyxl', nrows=1)
    return df
```

```python
dfs_antikeimena = [read_external_antikeimena(excel_filename, wn) for wn in worksheet_names]
```

```python
df_antikeimena = pd.concat(dfs_antikeimena, ignore_index=True, verify_integrity=True)
# df = pd.concat(dfs, ignore_index=True)
df_antikeimena.info()
```

```python
df_antikeimena.head()
```

```python
def read_eklektores_by_type(df: pd.DataFrame, charaktirismos: str) -> str:
    _list_int = df[df['Χαρακτη-ρισμός']==charaktirismos]['Κωδικός Χρήστη'].to_list()
    _list_str = [str(i) for i in _list_int]
    return '-'.join(_list_str)
```

```python
read_eklektores_by_type(dfs[0], 'ΙΔΙΟΥ')
```

```python
list(map(str, read_eklektores_by_type(dfs[0], 'ΙΔΙΟΥ')))
```

```python
print(read_eklektores_by_type(dfs[0], 'ΣΥΝΑΦΟΥΣ'))
```

```python
eklektores_idiou = [read_eklektores_by_type(df, 'ΙΔΙΟΥ') for df in dfs]
eklektores_synafous = [read_eklektores_by_type(df, 'ΣΥΝΑΦΟΥΣ') for df in dfs]
```

```python
df_antikeimena['Εξωτερικοί Ιδίου']  = eklektores_idiou
df_antikeimena['Εξωτερικοί Συναφούς']  = eklektores_synafous
df_antikeimena
```

```python
df_antikeimena.to_excel('antikeimena.xlsx')
```
